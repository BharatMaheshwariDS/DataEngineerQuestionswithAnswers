```python
# here we have total 428379 records
#answer to question two
#displaying the number of duplicates rows
total_rows=len(df)
df2=df.drop_duplicates(subset='cid', keep='first', inplace=False)
row_count=len(df2)
duplicate_rows=total_rows-row_count

print("Number of Duplcate Rows : ",duplicate_rows)
```
